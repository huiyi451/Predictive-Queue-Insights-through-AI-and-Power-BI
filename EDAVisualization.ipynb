```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your dataset (assuming it's in a CSV file)
df = pd.read_csv('queue_data.csv')
```

```python
# Create subplots for the three pie charts in a row
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Pie Chart for 'Scheduled Appointment'
appointment_counts = df['scheduled_appointment'].value_counts()
axes[0].pie(appointment_counts, labels=appointment_counts.index, autopct='%1.1f%%', startangle=90)
axes[0].set_title('Scheduled Appointment')
axes[0].axis('equal')

# Pie Chart for 'First Timer'
first_timer_counts = df['first_timer'].value_counts()
axes[1].pie(first_timer_counts, labels=first_timer_counts.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('First Timer')
axes[1].axis('equal')

# Pie Chart for 'Reason for Visit'
reason_counts = df['reason_visit'].value_counts()
axes[2].pie(reason_counts, labels=reason_counts.index, autopct='%1.1f%%', startangle=90)
axes[2].set_title('Reason for Visit')
axes[2].axis('equal')

# Adjust spacing between subplots
plt.tight_layout()

# Save the combined pie charts as a single image (e.g., PNG)
plt.savefig('visitinfo.png')

# Display the combined pie charts
plt.show()
```

```python
# Differentiate the data categories in the bar chart
plt.figure(figsize=(10, 6))

# Create a bar chart for 'Gender'
gender_counts = df['gender'].value_counts()
gender_counts_sorted = gender_counts.sort_index()  # Sort by gender
plt.bar(gender_counts_sorted.index, gender_counts_sorted.values, color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig('gender.png')

plt.show()
```

```python
# Create a bar chart for 'Number of Patients Waiting' (Count of Number Waiting)
plt.figure(figsize=(10, 6))
patient_waiting_counts = df['number_waiting'].value_counts().sort_index()
plt.bar(patient_waiting_counts.index, patient_waiting_counts.values, color='green')
plt.title('Number of Patients Waiting')
plt.xlabel('Number of Patients Waiting')
plt.ylabel('Count')
plt.xticks(range(1, max(df['number_waiting']) + 1))  # Set x-axis ticks from 1 to the maximum number of waiting patients
plt.savefig('numberwaiting.png')


plt.show()
```

```python
# Categorize ages into groups
age_bins = [0, 20, 40, 60, 80]  # Define the age group boundaries
age_labels = ['1-20', '21-40', '41-60', '61-80']  # Define the labels for the age groups
df['Age Group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

# Create a bar chart for 'Age Group'
plt.figure(figsize=(10, 6))
age_group_counts = df['Age Group'].value_counts().sort_index()
plt.bar(age_group_counts.index, age_group_counts.values, color='purple')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.savefig('age.png')

plt.show()
```

```python
# Create a bar chart for 'Specialty'
specialty_counts = df['specialty'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.bar(specialty_counts.index, specialty_counts.values, color='teal')
plt.title('Specialty Distribution')
plt.xlabel('Specialty')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('specialty.png')

plt.show()
```

```python
# Get the counts for each check-in time
checkin_time_counts = df['checkin_time'].value_counts().sort_index()

# Create a bar chart for check-in time counts
plt.figure(figsize=(12, 6))
plt.bar(checkin_time_counts.index, checkin_time_counts.values, color='blue')
plt.title('Check-in Time Counts')
plt.xlabel('Check-in Time')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('checkintime.png')

plt.show()
```

```python
# Get the counts for each check-in time
checkin_time_counts = df['checkout_time'].value_counts().sort_index()

# Create a bar chart for check-in time counts
plt.figure(figsize=(12, 6))
plt.bar(checkin_time_counts.index, checkin_time_counts.values, color='green')
plt.title('Check-out Time Counts')
plt.xlabel('Check-out Time')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('checkouttime.png')

plt.show()
```

```python
# Get the counts for each wait time
wait_time_counts = df['wait_time'].value_counts().sort_index()

# Create a bar chart for wait time counts
plt.figure(figsize=(12, 6))
plt.bar(wait_time_counts.index, wait_time_counts.values, color='green')
plt.title('Wait Time Counts')
plt.xlabel('Wait Time (minutes)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('waittime.png')

plt.show()
```
